```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

chemin_fichier_excel = "C:\\Users\\lecle\\Desktop\\ENSAE\\SX5Edec2021.xlsx"

df = pd.read_excel(chemin_fichier_excel)

df = pd.read_excel(chemin_fichier_excel, parse_dates=['Dates'])
df = df.rename(columns={'SX5E Index': 'Cours'})

# Calcul de la volatilit sur 60 jours
df['volatilite_60jours'] = df['Cours'].pct_change().rolling(window=60).std()

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Supposons que tu aies dj un DataFrame df avec la colonne 'Cours' et les colonnes calcules

# Convertir la colonne 'Dates' en type datetime si elle ne l'est pas dj
df['Dates'] = pd.to_datetime(df['Dates'])


# Conversion des dates en annes pour l'affichage
df['Anne'] = df['Dates'].dt.year

# Cration du graphique
plt.figure(figsize=(10, 6))
plt.plot(df['Dates'], df['volatilite_60jours'], label='Volatilit 60 jours')

# Ajout de labels et titre
plt.title('')
plt.xlabel('Date')
plt.ylabel('Volatilit 60 jours')

# Formater l'axe des x pour afficher uniquement les annes
plt.gca().xaxis.set_major_locator(mdates.YearLocator(4))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Ajout d'une lgende
plt.legend()

chemin_sauvegarde = 'C:\\Users\\lecle\\Desktop\\ENSAE\\Graphiques StatApp\\graphe_1.png'
plt.savefig(chemin_sauvegarde, dpi=300)
# Affichage du graphique
plt.show()
```
